# Chiffres Romains vers Entier

Les chiffres romains sont représentés par sept symboles différents :

| Symbole | Valeur |
|---------|--------|
| I       | 1      |
| V       | 5      |
| X       | 10     |
| L       | 50     |
| C       | 100    |
| D       | 500    |
| M       | 1000   |

Par exemple :
- 2 s'écrit `II` (deux fois 1).
- 12 s'écrit `XII` (soit `X + II`).
- 27 s'écrit `XXVII` (soit `XX + V + II`).

Les chiffres romains sont généralement écrits **du plus grand au plus petit** de gauche à droite. Cependant, dans certains cas, on utilise la **soustraction** au lieu de l'addition.

## Notation Soustractive

Le chiffre quatre ne s'écrit pas `IIII`, mais `IV`, car le `I` est placé avant le `V`, donc on le soustrait pour obtenir 4.

Ce principe s'applique aussi à d'autres chiffres :

Il existe six cas où l'on utilise la soustraction :

- `I` peut être placé avant `V` (5) et `X` (10) pour former **4** et **9**.
- `X` peut être placé avant `L` (50) et `C` (100) pour former **40** et **90**.
- `C` peut être placé avant `D` (500) et `M` (1000) pour former **400** et **900**.
